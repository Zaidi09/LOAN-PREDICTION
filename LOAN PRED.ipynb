# OUTLIER TREATMENT

```python
df_num=df.select_dtypes(np.number)
df_num
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
df_scale=pd.DataFrame(sc.fit_transform(df_num),columns=df_num.columns)
df_scale.reset_index(drop=True,inplace=True)
df_scale
```

```python
df_cat=df.drop(["Loan_ID","Loan_Status"],axis=1)
df_cat=df_cat.select_dtypes(np.object)
df_cat
```

```python
df_dum=pd.get_dummies(df_cat,drop_first=True)
df_dum
```

```python
final=pd.concat([df_scale,df_dum],axis=1)
final
```

# TRAIN TEST SPLIT

```python
x=final
y=df.Loan_Status
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.20,random_state=10)
```

# MODELING

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
```

```python
dc=DecisionTreeClassifier(random_state=10)
```

```python
mod=dc.fit(xtrain,ytrain)
y_pred=mod.predict(xtest)

print("REPORT",classification_report(ytest,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier,
```

```python
rc=RandomForestClassifier(random_state=10)
```

```python
mod=rc.fit(xtrain,ytrain)
y_pred=mod.predict(xtest)

print("REPORT",classification_report(ytest,y_pred))
```

```python
pd.DataFrame({"Features":x.columns,"IMP":mod.feature_importances_}).sort_values(by="IMP",ascending=False)
```
